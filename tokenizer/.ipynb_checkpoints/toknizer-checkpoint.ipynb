# create a list for text using regular expression

```python
import re
regular_text=re.split(r"[,.:;!_()?'\"]|--|\s",raw_text)
regular_text=[text.strip() for text in regular_text if text.strip()]
```

```python
print(regular_text[:100])
```

```python
# remove duplicates from the regular text
```

```python
print(len(regular_text))
```

```python
regular_text=sorted(list(set(regular_text)))
```

```python
len(regular_text)
```

```python
vocab={s:i for i,s in enumerate(regular_text)}
```

```python
for i,item in enumerate(vocab.items()):
    print(item)
    if i>500:
        break

    
```

```python
class simpleTokenizer:
    def __init__(self,vocab):
        self.token_to_id=vocab
        self.id_to_token={Id:token for token,Id in vocab.items() }
    def encoder(self,text:str)->list:
        list_text=re.split(r"[,.:;!_()?'\"]|--|\s",text)
        list_text=[word.strip() for word in list_text if word.strip()]
        
        ids=[]
        for word in list_text:
            ids.append(self.token_to_id[word])
        return ids
    def decoder(self,ids:list)->list:
        tokens=[]
        for i in ids:
            tokens.append(self.id_to_token[i])
        return tokens
        
```

```python
tokeniser=simpleTokenizer(vocab)
```

```python
tokeniser.encoder("on top of the universe")
```
