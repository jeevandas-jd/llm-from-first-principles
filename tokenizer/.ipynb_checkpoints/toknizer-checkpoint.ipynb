# building a simple tokenizer

```python
#step 1:extract text from a source here from a book
```

```python
from pypdf import PdfReader

text=PdfReader("../sample-book.pdf")
```

```python
print(type(text))
```

```python
print(text)
```

```python
page1=text.pages[0].extract_text()
print(page1)
```

```python
number_of_pages=len(text.pages)
with open("sample-book.txt","w") as f:
    for i in range(number_of_pages):
        f.write(text.pages[i].extract_text())

    
```

# now lets impliment vocabilary

```python
with open("sample-book.txt" ,encoding="utf-8") as f:
    raw_text=f.read()
print(len(raw_text))
```

```python
print(type(raw_text))
```

```python
print(raw_text[:500])
```

# create a list for text using regular expression

```python
import re
regular_text=re.split(r"[,.:;!_()?'\"]|--|\s",raw_text)
regular_text=[text.strip() for text in regular_text if text.strip()]
```

```python
print(regular_text[:100])
```

```python
# remove duplicates from the regular text
```

```python
print(len(regular_text))
```

```python
regular_text=sorted(set(regular_text))
```

```python
len(regular_text)
```

```python
vocab={s:i for i,s in enumerate(regular_text)}
```

```python
for i,item in enumerate(vocab.items()):
    print(item)
    if i>50:
        break

    
```
